```python
# plots for GFP

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3,figsize=(20, 5))

GFP = pd.read_csv('GFP.csv')

sns.lineplot(ax=axs[0],x='Time [ms]',y='l_soa_GFP',data=GFP,hue='Age',palette=['steelblue','orangered'],errorbar='se')
axs[0].set_title('VBM')
axs[0].set(ylabel='GFP [' + u'\u03bcV]')

sns.lineplot(ax=axs[1],x='Time [ms]',y='25_GFP',data=GFP,hue='Age',palette=['steelblue','orangered'],errorbar='se')
axs[1].set_title('CMot')
axs[1].set(ylabel='GFP [' + u'\u03bcV]')

sns.lineplot(ax=axs[2],x='Time [ms]',y='RT_GFP',data=GFP,hue='Age',palette=['steelblue','orangered'],errorbar='se')
axs[2].set_title('RT')
axs[1].set(ylabel='GFP [' + u'\u03bcV]')

```

```python
# plots for N1

fig, axs = plt.subplots(1, 3,figsize=(20, 5))

data = pd.read_csv('ERP_oneFeature.csv')
N1 = data[["Age","Subj","Time [ms]","l_soa_N1","25_N1","RT_N1"]]

axs[0].fill_between([100,300],1,-5.5,color='lightgray',alpha=0.5)
sns.lineplot(ax=axs[0],x='Time [ms]',y='l_soa_N1',data=N1,hue='Age',palette=['steelblue','orangered'],errorbar='se')

axs[1].fill_between([100,300],-3.5,4.2,color='lightgrey',alpha=0.5)
sns.lineplot(ax=axs[1],x='Time [ms]',y='25_N1',data=N1,hue='Age',palette=['steelblue','orangered'],errorbar='se')

axs[2].fill_between([100,300],-3.5,5,color='lightgrey',alpha=0.5)
sns.lineplot(ax=axs[2],x='Time [ms]',y='RT_N1',data=N1,hue='Age',palette=['steelblue','orangered'],errorbar='se')
```
